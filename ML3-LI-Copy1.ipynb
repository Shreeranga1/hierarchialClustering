 <span style='color:blue'>  

 1.Assume number of clusters(k)<br>
 2.Assume K centroids <br>
 3.Scale the data   ---StandardScaler,RobustScaler,MinMaxscaler<br>
 4.Compute distance between centroids and each observations <br>
 5.Assign the observation to the corresponding centroid group for which the distance is minimum.<br>
 6.Update the centroid by taking the mean of each centroid group columns <br>
 7.Stop the process when subsequent centroids are equal.<br>
 stop when difference between two centroids are very less.<br>

#### <span style='color:green'>#Validating Clusters: Silhoutte score <br>
1. Modelling both the distance between clusters and within the clusters<br>
2. [p(o)-q(o)/max(p(o),q(o))]<br>
3. computation is heavy <br>
4. p(o) point o within cluster p<br>
5. q(o) point o within cluster q<br>
6. Range -1 to +1 Silhoutte score <br>
7. more the positive more the seperatability between clusters<br> 
8. if its close to 0 then clusters are about to overlap<br>
9. if its negative point then overlap clusters(silhoutte score may be negative)<br>

    #### <span style='color:red'>#Quality of silhoutte <br>
1. all the silhoutte spectrum should cross average<br>
2. negative side should not be there <br>

## <span style='color:blue'> Hierarchical Clustering

1. <span style='color:green'>1.Bottom -Agglomerative <br>
2.  <span style='color:green'>2.Top Down --Divisive <br>
3.  <span style='color:green'>Bottom up<br>

1. all the samples as individual cluster<br>
2. Euclidean distance between each pair<br>
3. Find min distance <br>
4. Club the points(clusters) with min distances<br>
5. Update distance matrix using Linkage method<br>
6. Keep on till cluster will become 1<br>
    
    
    
1.<span style='color:red'>single</span> --min--min(AC,B)--clusters <br>
Pros:seperates non-elliptical shapes as long as gap between two clusters is not small.<br>
Cons:cannot seperate clusters properly if there is noise between clusters<br>

2.<span style='color:red'>complete </span>--max(AC,B)<br>
Pros: does well in seperating clusters if there is noise between clusters<br>
Cons:biased towards globular clusters.<br>

3.<span style='color:red'>avg</span>--avg(AC,B)<br>
Pros: does well in seperating clusters if noise between clusters<br>
Cons: biased towards globular clusters<br>

4.<span style='color:red'>ward </span>--Objective function--club cluster with other clusters--try all possibility grouping and 
 compute the error with min error combination would be taken <br>
 Pros:does good seperation of clusters if noise between clusters<br>
Cons:biased globular clusters<br>



<span style='color:red'>a. Dendogram </span>--identify optimum number clusters <br>
draw horizontal lines and vertical lines where maximum distance will remain same<br>

<span style='color:red'>b. Quality of cluster</span>-Cophenetic Coefficient --quality of linkage we used<br>
Cophenent index is a measure of the correlation between the distance of points in feature space and distance on the dendrogram.
If the distance between these points increases with the dendrogram distance between the clusters then the Cophenent index is closer to 1.<br>



<span style='color:green'>Hierarchical Clustering</span><br>
Space  Complexity<br>
O(N2) --N-number of data points<br>
store similarity matrix in RAM<br>

<span style='color:green'>Time Complexity<br>
O(N3)--N --number of data points<br>

<span style='color:green'>Limitations <br>
1.High space and Time complexity<br>
2.cannot be used for huge data<br>

```python
#Outlier Treatment
data_sc1=df_sc.copy()
for i in data_sc1.columns:
    q1=data_sc1[i].quantile(0.25)
    q3=data_sc1[i].quantile(0.75)
    iqr=q3-q1
    ub=q3 + 1.5*iqr
    lb=q1 - 1.5*iqr
    uc=data_sc1[i].quantile(0.99)
    lc=data_sc1[i].quantile(0.01)
    for ind1 in data_sc1[i].index:
        if data_sc1.loc[ind1, i] >ub:            
            data_sc1.loc[ind1, i] =uc
        if data_sc1.loc[ind1, i] < lb:
            data_sc1.loc[ind1, i] =lc
```

<span style='color:green'>PCA

1. Scale the data(Mean substraction)<br>
2. Compute the covariance matrix<br>
3. Find the eigen value and eigen vector for covariance matrix<br>
4. Order the eigen vector based on eigen values in descending order<br>
5. Perform dot vector between eigen vector and scaled data.<br>


a. The numbers of PCS will be equal to input features<br>
b. we may  not use all the PCs <br>
c. We may need to compute the numbers of PCs required to capture the variance<br>        of original data (initial PCs )<br>
d. All features are perpendicular<br>
e. corelation between PCs are 0 (independent of each other)<br>
f. Loading ---feature weightage in principal component <br>
g. PCA=L1*x1+L2*x2<br>
h. L1,L2,L3 --loadings (python pca.components_)<br>
i. visualize Pca components<br>
j. Build ML model with and without PCA components and compare performances<br> 


```python
from sklearn.decomposition import PCA
pca=PCA()
pcadata=pca.fit_transform(df_sc)
```

```python
### 2.  Apply PCA and compute all the possible principle components(PCs). 
#   How many PCs are required to reproduce the 95% charecteristics of original data. 
#   Plot it with appropriate diagram. Also print the top 5 eigen vectors (5 marks)

from sklearn.decomposition import PCA 
pca = PCA(n_components = df_sc.shape[1])
pca_data = pca.fit_transform(df_sc)
exp_var_ratio= pca.explained_variance_ratio_
exp_var_ratio.round(3)

cum_var=exp_var_ratio[0]
itr=2 # defined as two as first pc1 variance defined outside the loop
for j in exp_var_ratio[1:]:
    cum_var=cum_var+j
    if cum_var >= 0.95:
        break
    itr=itr+1

print('The number of principle components capturing 95 percent variation in data is : ',itr,' Variance explained is ', cum_var)
```
